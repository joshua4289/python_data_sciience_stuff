```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
points =np.arange(-5,5,0.01)
```

```python
dx,dy = np.meshgrid(points,points)
```

```python
z = (np.sin(dx)+ np.sin(dy))
```

```python
z
```

```python
plt.imshow(z)
plt.colorbar()
plt.title('Plot of sin(x) + sin(y)')
```
